## Gradient descent algorithm implemetation

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.arange(11)
Y=(X-5)**2;
plt.style.use("seaborn")
error=[]
x=0
lr=0.1
plt.plot(X,Y)
for i in range(50):
    grad=2*(x-5) # just differantiation
    x=x-lr*grad
    e=(x-5)**2
    error.append(e)
    plt.xlabel("Variation of x")
    plt.ylabel("Variation of y")
    plt.scatter(x,e)
print("Local Minima at x=",x) # it is near to global minima as x=5
plt.show()
```

```python
plt.plot(error)
print("error in all iteration      ",error)
plt.show()
```

i) step size(n)
ii) Direction
