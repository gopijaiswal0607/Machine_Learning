## Surface plot / Data Visualisation
- Visualise loss function in machine learing & Deep learning
- Visualise state or state value Function in Reforcement Learning

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
a=np.arange(-1,1,0.02)
b=a
a,b=np.meshgrid(a,b)
```

```python
fig=plt.figure()
axes=fig.gca(projection='3d')
axes.plot_surface(a,b,a**2+b**2,cmap='rainbow')
plt.title('3D surface')
plt.show()
```

```python
fig=plt.figure()
axes=fig.gca(projection='3d')
plt.contour(a,b,a**2+b**2,cmap='rainbow')
# we can give label of each axes
plt.show()
```
