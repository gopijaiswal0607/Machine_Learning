## Closed From Solution In Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import make_regression
```

```python
# Genrate dataset
X,Y=make_regression(n_samples=400,n_features=1,n_informative=1,noise=1.8,random_state=11)
```

```python
print(X.shape) # it is a matrix 
print(Y.shape) # it is a vector so reshape it as a matrix
Y=Y.reshape(-1,1)# reshape the Y vector
Y.shape
```

```python
plt.plot(X,Y)
```

```python
# Noramlised our dataset
print(X.mean(),X.std())
```

```python
X=(X-X.mean())/X.std()
# Visualise our normalised data
plt.plot(X,Y)
```

```python
plt.figure()
plt.scatter(X,Y)
plt.title("Normalized Data")
plt.show()
```

```python
# one dummy X0 is need so grenerate it
ones=np.ones((X.shape[0],1))
X_=np.hstack((X,ones))
print(X_.shape)
print(X_[ :5,])
```

```python

def predict(X,theta):  # 0=(X.traspose x X)(X.traspose x Y)
    return np.dot(X,theta)
def getThetaClosedFrom(X,Y):
    Y=np.mat(Y)
    first_part=np.dot(X.T,X)
    second_part=np.dot(X.T,Y)
    theta=np.linalg.pinv(first_part)*second_part
    return theta
    
    
```

```python
theta=getThetaClosedFrom(X_,Y)
theta=np.array(theta)
```

```python
plt.style.use("seaborn")
plt.figure()
plt.scatter(X,Y)
plt.scatter(X,predict(X_,theta),color='red',label="Predict")
plt.legend()
plt.show()
```
